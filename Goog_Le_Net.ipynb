```python
from tensorflow.keras.preprocessing import image
import tensorflow as tf
from tensorflow.keras.applications.inception_v3 import *
import numpy as np 
from matplotlib import pyplot as plt 
from tensorflow.keras.utils import plot_model
from PIL import Image
```

## Importiamo il modello preaddestrato InceptionV3 di keras 

```python
model = InceptionV3(weights = 'imagenet')
```

```python
model.summary()
```

## Importiamo l'immagine sul quale vogliamo avere le predizioni del modello 

```python
img = image.load_img(r'C:\Users\claud\Desktop\Python\MASTER CLASS MACHINE LEARNING\moduli\3 modulo\office.jpg', target_size=(299, 299))

plt.imshow(img)
plt.show()
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
print(x.shape)
```

## Addestramento del modello 

```python
y = model.predict(x)
decoded_predictions = decode_predictions(y)
print(decoded_predictions)
```

## Visualiziamo i risultati di addestramento del modello 

```python
for index, res in enumerate(decoded_predictions[0]):
    print('{}. {}: {:.3f}%'.format(index + 1, res[1], 100 * res[2]))

print('---')

for index, res in enumerate(decode_predictions(y, top=20)[0][5:]):
    print('{}. {}: {:.3f}%'.format(index + 6, res[1], 100 * res[2]))
```
